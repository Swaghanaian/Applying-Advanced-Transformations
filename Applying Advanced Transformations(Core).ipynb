# Superhero Info

```python
df_info = pd.read_csv('superhero_info - superhero_info.csv')
df_info.head()

```

```python
df_info.info()
```

```python
# Exploring existing format with a few examples
df_info['Hero|Publisher'].head(2)

```

```python
## adding expand=True
df_info['Hero|Publisher'].str.split('|',expand=True)

```

```python
## save the 2 new columns into the dataframe
df_info[['Hero','Publisher']] = df_info['Hero|Publisher'].str.split('|',expand=True)
df_info.head(2)
```

```python
## drop the original column 
df_info = df_info.drop(columns=['Hero|Publisher'])
df_info.head(2)
```

```python
## examining a single value from the measurement col
measurements = df_info.loc[0,"Measurements"]
print(type(measurements))
measurements
```

```python
#fixing str dict col with json
#json.loads(measurements)



#fixing error: Expecting property name enclosed in double quotes: line 1 column 2 (char 1)
measurements = measurements.replace("'",'"')
measurements

```

```python
## now using json.loads
fixed_measurements = json.loads(measurements)
print(type(fixed_measurements))
fixed_measurements
```

```python
## use .str.replace to replace all single quotes in the dataframe for measurements column
df_info['Measurements'] = df_info['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
df_info['Measurements'].head()

```

```python
## checking a single value after transformation
test_transformation = df_info.loc[0, 'Measurements']
print(type(test_transformation))
test_transformation
```

```python
#Unpacking column of dictionaries into separate columns
height_weight = df_info['Measurements'].apply(pd.Series)
height_weight

```

```python
df_info
```

```python
# concat height_weight with original dataframe
df_info = pd.concat((df_info, height_weight), axis = 1)
df_info.head(2)
```

```python
#dropping measurements col
df_info = df_info.drop(columns=['Measurements'])
```

```python
#Verifying dataframe looks good
df_info
```

# Superhero Powers

```python
df_powers = pd.read_csv('superhero_powers - superhero_powers.csv')
df_powers.head()
```

```python
df_powers['Powers'].value_counts()

```

```python
df_powers.loc[0,'Powers']
```

```python
# splitting powers into columns to create a cleaner list
df_powers['Powers_split'] = df_powers['Powers'].str.split(',',expand =False)
```

```python
#save columns into the dataframe
df_powers[['Powers']]
```

```python
## exploding the column of lists
exploded = df_powers.explode('Powers_split')

exploded
```

```python
# creating columns to explode into
cols = exploded['Powers_split'].dropna().unique()
cols
```

```python
# loop through cols to one hot encode powers
for unique_power in cols:
    df_powers[unique_power] = df_powers['Powers'].str.contains(unique_power)
df_powers.head()
```

```python
#drop orignal powers column and powers split column
df_powers = df_powers.drop(columns = ['Powers','Powers_split'])
df_powers.head(5)
```

```python
#concantenate powers df and info df
combined_powers = [df_info, df_powers]
  
result = pd.concat(combined_powers)
display(result)
```
